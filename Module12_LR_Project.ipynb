```python
# Task 22:
# Obtain X_train and y_train by using drop option. Note that "charges" is your y_train. 
X_train = train_data.drop('charges', axis=1)
y_train = train_data['charges']
```

```python
# Task 23:
# Obtain X_test and y_test by using drop option
X_test = test_data.drop('charges', axis=1)
y_test = test_data['charges']
```

```python
X_test.head()
```

```python
# Task 24:
# Obtain categorical features i.e. sex, children, smoker and region into a separate variable DF cat_feats

cat_feats = X_train[['sex', 'children', 'smoker', 'region']]
```

```python
cat_feats.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()

cat_feats = X_train[['sex', 'children', 'smoker', 'region']]
cat_feats_hot = encoder.fit_transform(cat_feats)
cat_feats_hot
```

```python
cat_feats_hot.toarray()
```

```python
# Task 25: 
# Build a categorical feature OneHotEncoder pipeline to produce cat_feats_preprocessed
from sklearn.pipeline import Pipeline
cat_feats_pipe = Pipeline([
    
    ('encoder', OneHotEncoder())
])
cat_feats_preprocessed = cat_feats_pipe.fit_transform(cat_feats)
cat_feats_preprocessed
```

```python
cat_feats_preprocessed.toarray()
```

```python
# Task 26:
# Obtain numerical features i.e. age and bmi into a separate variable DF num_feats

num_feats = X_train[['age', 'bmi']]
num_feats.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Task 27: 
# Build a numerical Standardization pipeline to produce num_feats_preprocessed
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler 

num_feats_pipe = Pipeline([
    ('scalar', StandardScaler())
])
num_feats_preprocessed = num_feats_pipe.fit_transform(num_feats)
num_feats_preprocessed
```

```python
# Task 28:
# Build Final pipeline by combing above two pipelines. Use ColumnTransformer
from sklearn.compose import ColumnTransformer

# The transformer requires list of features
num_list = list(num_feats)
cat_list = list(cat_feats)

final_pipe = ColumnTransformer([
    ('num', num_feats_pipe, num_list),
    ('cat', cat_feats_pipe, cat_list)
])
X_train_preprocessed = final_pipe.fit_transform(X_train)
X_train_preprocessed
```

```python
# Task 29:
# Create a linear regression model

from sklearn.linear_model import LinearRegression
reg_model= LinearRegression()
reg_model
```

```python
# Task 30:
# Fit it with X_train_preprocessed and y_train
reg_model.fit(X_train_preprocessed, y_train)
```

```python
# Task 31:
# Determine its coefficient and intercept
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
# Task 32: 
# Determine y_train_pred i.e. predicted outputs or charges for train data

y_train_pred = reg_model.predict(X_train_preprocessed)
```

```python
# Task 33:
# Create a distplot of y_train and y_train_pred i.e. actual charges and predicted charges 
plt.figure(figsize=[12,8])
sns.distplot(y_train)
sns.distplot(y_train_pred)
```

```python
# Task 34:
# Determine RMSE for predictions of train data
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_train_pred, y_train)
RMSE = np.sqrt(MSE)
RMSE
```

```python
# Task 35:
# Obtain X_test_preprocessed from X_test by applying the pipeline
X_test_preprocessed = final_pipe.fit_transform(X_test)
X_test_preprocessed
```

```python
# Task 36:
# Obtain y_test_pred i.e. predicted values for test data
y_test_pred = reg_model.predict(X_test_preprocessed)
MSE_test = mean_squared_error(y_test_pred, y_test)
RMSE_test = np.sqrt(MSE_test)
RMSE_test
```

```python
# Task 37:
# Determine the accuracy of the training data predictions
reg_model.score(X_train_preprocessed, y_train)
```

```python
# Task 38: 
# Determine the accuracy of the test data predictions
reg_model.score(X_test_preprocessed, y_test)
```
